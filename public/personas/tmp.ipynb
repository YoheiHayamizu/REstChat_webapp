```python
import json
from pprint import pprint
import pandas as pd
from glob import glob
```

```python
with open('./all_personas.json') as f:
    persona_data = json.load(f)
persona_df = pd.DataFrame(persona_data)
# Change keys to lowercase
persona_df.columns = [x.lower() for x in persona_df.columns]
# Change marital status to marital_status
persona_df.rename(columns={"marital status": "marital_status"}, inplace=True)
# Add id column
persona_df["id"] = persona_df.index.astype(str).str.zfill(3)
persona_df.to_json("all_personas.json", orient="records")
persona_df
```
